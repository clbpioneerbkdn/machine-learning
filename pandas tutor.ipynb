```python
import pandas as pd

df3 = pd.read_csv("avocado.csv")
df3.head()
```

```python
df4 = df3.copy()[df3['type'] == "organic"]
df4.sort_values(by="Date",ascending=True,inplace=True)
#sắp xếp giá trị của cột "Date" theo thứ tự tăng dần
df4.head(100)
```

```python
#khởi tạo dataframe :
region_df = pd.DataFrame()
for region in df4['region'].unique():
#     print(region)
    graph_df = df4[df4["region"] == region]
    #lấy ra  dataframe của mỗi khu vực
    #[df4["region"]] : lấy toàn bộ cột của bảng df4 có điều kiện thỏa mãn
    graph_df.set_index("Date",inplace=True)
    # set index là cột Date
    graph_df.sort_index(inplace=True)
    #sắp xếp cột index mặc định là ascending là True nên tăng dần 
    graph_df[f'{region}_price25mean'] = graph_df['AveragePrice'].rolling(25).mean()
    # tạo 1 cột mới có tên là region_price25mean lấy giá trị từ 'AvaragePrice' sau khi lấy trung bình từ mỗi 25 số 
    if region_df.empty:
        region_df = graph_df[[f'{region}_price25mean']]
        #thêm cột region_price25mean vào DataFrame mới
    else:
        region_df = region_df.join(graph_df[[f'{region}_price25mean']])
        #tiếp tục thêm giá trị của các region khác vào 
region_df.tail(100)
```
